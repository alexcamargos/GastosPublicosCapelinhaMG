## Dados Abertos

### Conceito
Segundo a [Open Knowledge Foundation no Brasil - OKBR](https://ok.org.br), _“dados são abertos quando qualquer pessoa pode livremente *usá-los*, *reutilizá-los* e *redistribuí-los*, estando sujeito a, no máximo, a exigência de creditar a sua autoria e compartilhar pela mesma licença”_. Quando os dados são produzidos, coletados ou custodiados por autoridades públicas e disponibilizados em formato aberto, nesse caso temos dados abertos governamentais. Ainda segundo a OKBR, dados abertos também são pautados por três leis e oito princípios.

As três leis:
- I - Se o dado não pode ser encontrado e indexado na Web, ele não existe;
- II - Se não estiver aberto e disponível em formato compreensível por máquina, ele não pode ser reaproveitado; e
- III - Se algum dispositivo legal não permitir sua replicação, ele não é útil.

Os oito princípios:
1) *Completos*: todos os dados públicos são disponibilizados. Dados são informações eletronicamente gravadas, incluindo, mas não se limitando a, documentos, bancos de dados, transcrições e gravações audiovisuais. Dados públicos são dados que não estão sujeitos a limitações válidas de privacidade, segurança ou controle de acesso, reguladas por estatutos.
2) *Primários*: os dados são publicados na forma coletada na fonte, com a mais fina granularidade possível, e não de forma agregada ou transformada.
3) *Atuais*: os dados são disponibilizados o quão rapidamente seja necessário para preservar o seu valor.
4) *Acessíveis*: os dados são disponibilizados para o público mais amplo possível e para os propósitos mais variados possíveis.
5) *Processáveis por máquina*: os dados são razoavelmente estruturados para possibilitar o seu processamento automatizado.
6) *Acesso não discriminatório*: os dados estão disponíveis a todos, sem que seja necessária identificação ou registro.
7) *Formatos não proprietários*: os dados estão disponíveis em um formato sobre o qual nenhum ente tenha controle exclusivo.
8) *Livres de licenças*: os dados não estão sujeitos a regulações de direitos autorais, marcas, patentes ou segredo industrial. Restrições razoáveis de privacidade, segurança e controle de acesso podem ser permitidas na forma regulada por estatutos.

O objetivo, além da promoção da transparência governamental, é fornecer insumos aos cidadãos e à sociedade civil para serem desenvolvidas ferramentas que proporcionem uma percepção mais efetiva das atividades e dos gastos do governo e contribuam para a melhoria da gestão pública, o estímulo ao controle e participação social, a geração de emprego e renda e o fomento à inovação tecnológica.

No Brasil, conforme o [Manual dos Dados Abertos: Governo](https://www.w3c.br/pub/Materiais/PublicacoesW3C/Manual_Dados_Abertos_WEB.pdf), dados abertos governamentais são particularmente importantes devido à quantidade e centralidade dos dados que coleta e, também, por tais dados serem públicos, um direito garantido no art. 5º da Constituição Federal brasileira.

## Limpeza de Dados (Data Cleaning)
Para podermos conduzir uma análise verdadeiramente confiável, é imprescindível trabalhar com o propósito de aprimorar nossos dados à sua melhor versão possível. O objetivo do data cleaning, ou limpeza de dados, é justamente maximizar a qualidade e a precisão dos dados, tornando-os mais relevantes e úteis para análise e tomada de decisão. Esse processo pode abranger a eliminação de registros duplicados, complementação de valores ausentes, correção de erros de digitação, além da verificação e manutenção da consistência e conformidade dos dados em relação a padrões específicos.

### Convertendo o tipo dos valores

Para aprimorarmos nosso entendimento do DataFrame, vamos examinar como o _Pandas_ identificou os tipos de cada dado durante o processo de importação.

```python
# Entendo os tipos dos dados para cada coluna.
payment_df.dtypes
```

O resultado da análise será tão preciso quanto a precisão da fonte de dados. ## Análise dos Dados (Data Analysis)

Até agora, já realizamos um trabalho considerável, no entanto, estávamos principalmente focados no tratamento dos dados para simplificar o processo de análise e criação de gráficos. A partir deste momento, vamos direcionar nossos esforços para responder algumas questões utilizando a base de dados que temos em mãos.

Iniciaremos nossa análise identificando o valor total despendido pelo município ao longo do ano de 2022, bem como delineando quais foram os vinte principais credores.

```python
total_expenses_2022 = payment_df['vlr_pag_fonte'].sum()
total_expenses_2022_formatted = locale.format_string('%.2f', total_expenses_2022, grouping=True)
print(f'O município de Capelinha/MG realizou pagamentos no valor total de R${total_expenses_2022_formatted} em 2022.')
```

Embora a informação referente ao valor total gasto pelo município durante o ano seja interessante, ela não fornece muitos insights relevantes. Por isso, vamos analisar quem são os principais credores municipais.

```python
# Agrupando os pagamentos realizados para cada credor ao longo do ano.
creditors = payment_df.groupby('nom_credor').agg(
    {'vlr_pag_fonte': 'sum', 'vlr_ret_fonte': 'sum', 'vlr_ant_fonte': 'sum', 'vlr_anu_fonte': 'sum'}
)
```

```python
# Quais são os 20 maiores credores.
creditors.nlargest(20, 'vlr_pag_fonte')
```

_“Houston, we have a problem”_. O fato de nosso principal credor não ter um nome definido é bastante estranho, especialmente porque a administração pública é obrigada a informar o destinatário de um pagamento. Vamos analisar de perto os lançamentos referentes a esse credor não identificado, denominado *NAO INFORMADO*.

```python
# Para a coluna nom_credor filtre apenas os campos que contenha o lançamento 'NAO INFORMADO'.
payment_df.query('nom_credor == "NAO INFORMADO"')['dsc_pagamento'].sample(10, random_state=RANDOM_SEED)
```

Agora faz sentido. Esse pagamento se refere à quitação da folha de pagamento do município ao longo do ano, portanto, não é direcionado a um credor específico. Para melhorar o processo de análise, vamos renomear esse lançamento para FOLHA DE PAGAMENTO DOS SERVIDORES(AS) MUNICIPAIS

```python
# Trocando a informação 'NAO INFORMADO' da coluna nom_credor por FOLHA DE PAGAMENTO DOS SERVIDORES(AS) MUNICIPAIS.
payment_df['nom_credor'] = (payment_df['nom_credor']
                            .apply(
    lambda x: 'FOLHA DE PAGAMENTO DOS SERVIDORES(AS) MUNICIPAIS' if x == 'NAO INFORMADO' else x))
```

Vamos totalizar os gastos por cada credor novamente e, desta vez, também vamos totalizar as outras colunas que contêm valores monetários.

```python
creditors_adjustment = (payment_df.groupby('nom_credor')
                        .agg({'vlr_pag_fonte': 'sum',
                              'vlr_ret_fonte': 'sum',
                              'vlr_ant_fonte': 'sum',
                              'vlr_anu_fonte': 'sum'}))
creditors_adjustment.nlargest(20, 'vlr_pag_fonte')
```

```python
# Relembrando quais são os órgãos municipais ordenadores de gastos.
agency_df
```

Até o momento, estamos analisando os gastos totais do município. No entanto, a base de dados nos permite separar os gastos pelos dois principais órgãos municipais. Portanto, vamos analisar os pagamentos realizados pela Câmara de Vereadores e pela Prefeitura Municipal.

```python
# De acordo com nosso dicionário de dados na coluna seq_orgao_empenho o valor
# 371 corresponde a Câmara de Vereadores e o valor 372 corresponde a Prefeitura Municipal.
payment_councilors_camera_df = payment_df.query('seq_orgao_empenho == 371')
payment_city_hall_df = payment_df.query('seq_orgao_empenho == 372')
```

```python
# Totalizando os pagamentos realizados pela Câmara de Vereadores.
total_expenses_councilors_camera = payment_councilors_camera_df['vlr_pag_fonte'].sum()
total_expenses_councilors_camera_formatted = locale.format_string('%.2f',
                                                                  total_expenses_councilors_camera,
                                                                  grouping=True)
print(
    f'A Câmara de Vereadores de Capelinha/MG realizou pagamentos no valor total de R${total_expenses_councilors_camera_formatted} em 2022.')
```

```python
# Totalizando os pagamentos realizados pela Prefeitura Municipal.
total_expenses_city_hall = payment_city_hall_df['vlr_pag_fonte'].sum()
total_expenses_city_hall_formatted = locale.format_string('%.2f',
                                                          total_expenses_city_hall,
                                                          grouping=True)
print(
    f'A Prefeitura Municipal de Capelinha/MG realizou pagamentos no valor total de R${total_expenses_city_hall_formatted} em 2022.')
```

Quem são os principais credores da Câmara de Vereadores?

```python
# Agrupando os pagamentos por credor e somando os valores pagos.
creditors_councilors_camera = (payment_councilors_camera_df
                               .groupby('nom_credor')
                               .agg({'vlr_pag_fonte': 'sum'}))
creditors_councilors_camera.nlargest(20, 'vlr_pag_fonte')
```

Neste momento da análise, nos deparamos com uma informação que precisa ser examinada com mais detalhes. A Câmara de Vereadores pagou ao Tribunal Regional do Trabalho da 03ª Região, Vara do Trabalho de Guanhães o montante de R$1.582.708,46. Precisamos avaliar esse pagamento com mais detalhes.

```python
# Mostrando os pagamentos realizados para o credor
# TRIBUNAL REGIONAL DO TRABALHO DA 03ª REGIAO VARA DO TRABALHO DE GUANHAES.
payment_councilors_camera_df.query(
    'nom_credor == "TRIBUNAL REGIONAL DO TRABALHO DA 03ª REGIAO VARA DO TRABALHO DE GUANHAES"')[
    ['dat_liquidacao', 'num_doc_credor', 'dsc_pagamento', 'vlr_pag_fonte']].sample(10,
                                                                                   random_state=RANDOM_SEED).sort_values(
    by='dat_liquidacao')
```

Podemos concluir que esses pagamentos se referem à quitação da folha de pagamento dos servidores da Câmara de Vereadores, e não têm, a princípio, nenhuma relação com pagamentos para o Tribunal Regional do Trabalho da 03ª Região, Vara do Trabalho de Guanhães. Para evitar confusões, vamos renomear esse credor para “Folha de Pagamento dos Servidores(as) da Câmara de Vereadores”.

Como os pagamentos se comportam ao longo dos meses?

```python
# Criando uma grade de subplots 4x3.
fig, axs = plt.subplots(4, 3, figsize=(14, 10))

# Inteirando sobre todos os meses.
for month in range(1, 13):
    # Determina em qual linha do subplot vamos inserir o gráfico.
    row = (month - 1) // 3
    # Determina em qual coluna do subplot vamos inserir o gráfico.
    col = (month - 1) % 3
    ax = axs[row, col]  # Seleciona o subplot correto.
    # Cria o histograma no subplot correto.
    sns.histplot(x=payment_df.query(f'dat_pagamento.dt.month == {month}')['dat_pagamento'].dt.day,
                 data=payment_df.query(f'dat_pagamento.dt.month == {month}'),
                 bins=range(1, 32),
                 kde=True,
                 ax=ax)
    # Define o título para cada elemento do subplot.
    ax.set_title(f'{payment_month.index[month - 1].capitalize()}')
    # Remove o label do eixo X e Y.
    ax.set_xlabel('')
    ax.set_ylabel('')
    # Ajusta os ticks do eixo x para mostrar os dias a cada 2.
    ax.set_xticks(range(1, 32, 2))

# Define um título para a figura.
fig.suptitle('Pagamentos realizados por dia do mês', fontsize=16)

# Ajusta o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

Ao analisarmos a distribuição dos pagamentos ao longo dos meses, identificamos um padrão que se alinha às nossas expectativas. Considerando que a folha de pagamento representa a maior despesa do município, é natural que a maior concentração de gastos ocorra no início de cada mês ou, em alguns cenários, no último dia do mês. Contudo, vale ressaltar uma exceção: em novembro, os pagamentos tendem a estar mais concentrados no meio do mês. Isso pode ser explicado pelo fato de que, em novembro, os servidores públicos recebem a primeira parcela do 13º salário, o que pode ter impactado a distribuição dos pagamentos. Para aprimorar nossa compreensão, vamos investigar o comportamento do valor total dos pagamentos efetuados ao longo dos meses.

```python
# Criando uma grade de subplots 4x3.
fig, axs = plt.subplots(4, 3, figsize=(14, 10))

# Inteirando sobre todos os meses.
for month in range(1, 13):
    # Determina em qual linha do subplot vamos inserir o gráfico.
    row = (month - 1) // 3
    # Determina em qual coluna do subplot vamos inserir o gráfico.
    col = (month - 1) % 3
    ax = axs[row, col]  # Seleciona o subplot correto.
    # Filtra os dados para o mês atual e agrupa por dia do mês, somando os valores pagos.
    monthly_data = payment_df.query(f'dat_pagamento.dt.month == {month}').groupby(payment_df['dat_pagamento'].dt.day)[
        'vlr_pag_fonte'].sum()
    # Cria o gráfico de barras no subplot correto.
    sns.barplot(x=monthly_data.index, y=monthly_data.values, ax=ax)
    # Define o título para cada elemento do subplot.
    ax.set_title(f'{payment_month.index[month - 1].capitalize()}')
    # Remove os label dos eixos X e Y.
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_yticklabels('')

# Ajusta o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```
